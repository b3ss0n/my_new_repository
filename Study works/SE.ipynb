```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Воспроизведение исходных данных
rng = np.random.default_rng(seed=42)
matrix = rng.normal(loc=0.0, scale=1.0, size=(100, 100))
means = np.mean(matrix, axis=1)

stds = np.std(matrix, axis=1)

# Корректный расчет SE (стандартное отклонение распределения средних)
grand_mean = np.mean(means)
variance_of_means = np.sum((means - grand_mean)**2) / (100 - 1)
se = np.sqrt(variance_of_means)
print(se)
# 2. Построение графика
fig, ax = plt.subplots(figsize=(12, 6))
x = np.arange(1, 101)

# Точки средних значений с усами (в пределах $\pm$ 1 SD для каждой выборки)
ax.errorbar(x, means, yerr=stds, fmt='o', color='#1f77b4', ecolor='lightgray', 
            elinewidth=1.5, capsize=2, label='Среднее выборки $\pm$ SD данных')

# Линия общего среднего
ax.axhline(grand_mean, color='red', linestyle='--', linewidth=1.5,
           label=f'Общее среднее ({grand_mean:.4f})')

# Интервал стандартной ошибки средних (SE)
ax.axhspan(grand_mean - se, grand_mean + se, color='red', alpha=0.15, 
           label=f'Интервал $\pm$ SE средних ($\pm$ {se:.4f})')

# Оформление осей
ax.set_xlabel('Номер выборки', fontsize=11)
ax.set_ylabel('Значение', fontsize=11)
ax.set_title('Выборочные средние значения и их отклонения', fontsize=13, fontweight='bold')
ax.legend(loc='upper right')
ax.grid(True, linestyle=':', alpha=0.6)

plt.tight_layout()
plt.savefig('means_error_bars.png', dpi=300)
plt.show()
```
