Закон больших чисел и центральная предельная теорема на практике

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 

n_trials = 10000
elements = [0, 1]
probabilities = [0.2, 0.8]

data = np.random.choice(elements, size=n_trials, p=probabilities)
cumulative_average = np.cumsum(data) / np.arange(1, n_trials + 1)


matrix_of_experiments = np.random.choice(elements, size=(1000, n_trials), p=probabilities)
# Считаем среднее по каждому из 1000 экспериментов (по строкам, axis=1)
sample_means = np.mean(matrix_of_experiments, axis=1)

fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# --- График 1: ЗБЧ (Сходимость одного эксперимента) ---
axes[0].plot(cumulative_average, color='blue', alpha=0.7)
axes[0].axhline(y=0.8, color='red', linestyle='--', linewidth=2)
axes[0].set_ylim(0.5, 1.0)
axes[0].grid(True, alpha=0.3)
axes[0].set_title('Закон больших чисел (Сходимость)')
axes[0].set_xlabel('Итерации')
axes[0].set_ylabel('Текущее среднее')

# --- График 2: Распределение выборки ---

sns.histplot(data, bins=2, ax=axes[1], color='#66b3ff', edgecolor='black')
axes[1].set_xticks([0, 1])
axes[1].set_title('Распределение выборки')
axes[1].set_xlabel('Значение')
axes[1].set_ylabel('Количество значений')

# --- График 3: Плотность распределения средних (ЦПТ) ---
sns.histplot(sample_means, kde=True, ax=axes[2], color='purple', stat='density', bins=30)
axes[2].axvline(np.mean(sample_means), color='black', linestyle='dashed', linewidth=2)
axes[2].set_title('Плотность выборочных средних (ЦПТ)')
axes[2].set_xlabel('Выборочное среднее')
axes[2].set_ylabel('Плотность')

fig.tight_layout()
plt.show()
```
